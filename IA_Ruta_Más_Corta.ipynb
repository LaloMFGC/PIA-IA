#Breadth First Search

```python
def resolver(s,e):
  cola=queue.Queue(0)
  cola.put(s)
  
  visited=np.full(n,False)
  visited[s]=True

  last=np.full(n,None)

  while cola.empty()==False:
    nodo=cola.get()
    vecinos=[i for i in grafo.get(nodo)]

    for sig in vecinos:
      if sig==e: 
        last[sig]=nodo
        return last
      if visited[sig]==False:
        cola.put(sig)
        visited[sig]=True
        last[sig]=nodo

  return
```

```python
def makepath(s,e,last):
  i=e
  path=[]
  while last[i]!=None:
    path.append(i)
    i=last[i]
  path.append(i)
  path.reverse()
  return path
```

```python
n=len(grafo)

def bfs(s,e):
  anterior=resolver(s,e)
  if len(anterior):
    print('Camino del nodo inicial', s, 'al nodo final', e, makepath(s,e,anterior))
    n
  else:
    print('No hay un camino entre ese par de nodos')

```

```python
# Tomar el tiempo para 10 ejecuciones
sumatime = 0
for i in range(10):
  tiempo_transcurrido = 0
  tiempo_inicio = time()
  bfs(0,8) 
  tiempo_transcurrido = (time() - tiempo_inicio)
  print(tiempo_transcurrido)
  sumatime += tiempo_transcurrido
print("Promedio: ", sumatime/10)
```
